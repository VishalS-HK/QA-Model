```python
from transformers import AutoModelForQuestionAnswering, pipeline, AutoTokenizer
```

```python
modelName = "deepset/roberta-base-squad2"
```

```python
model = AutoModelForQuestionAnswering.from_pretrained(modelName)
tokenizer = AutoTokenizer.from_pretrained(modelName)
```

```python
nlp = pipeline('question-answering', model = modelName, tokenizer = modelName)
```

```python
filePath = "/content/walden.txt"
```

```python
with open(filePath, 'r') as f:
  contents = f.read()
```

```python
QA_input = {
    'question': 'Where is Walden Pond located',
    'context': contents
}
```

```python
res = nlp(QA_input)
```
